# 사각형 넓이 구하기

```python
# 사각형 넓이를 구하는 클래스 완성!
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        base, height = map(int, input('및변,높이를 입력하세요. 예시 : 및변,높이\n >>>').split(','))
        area = base * height
        result = '평행사변형 넓이는 : ' + str(area)
        return result

    def trape(self):
        top, base, height = map(int,input('윗변,및변,높이를 입력하세요. 예시 : 윗변,및변,높이\n >>>').split(','))
        area = (top + base) * height / 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

a = Square()  # 객체 생성 & 2, 3을 각각 입력해 봅시다.
```

```python
# 평형사변형의 넓이
a.par()  # 3,2를 입력했을 때 넓이가 6이 나와야 합니다.
```

```python
# 사다리꼴의 넓이
a.trape()  # 3,2,1 을 입력했을 때 넓이가 2.5이 나와야 합니다.
```

# 키오스크 만들기

```python
# 문제 3-3. 클래스 업데이트  
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요! 

menu_temp = ['americano', 'latte', 'mocha', 'choco_latte'] 
price_temp = [2000, 3000, 3000, 3000] 


class Kiosk: 
    def __init__(self):
        self.menu_temp = menu_temp
        self.price_temp = price_temp
        self.order_menu = []
        self.order_price = []
        self.price_sum = 0

    def menu_print(self):
        for i in range(len(self.menu_temp)):
            print(i + 1, self.menu_temp[i], ' : ', self.price_temp[i], '원')

    def menu_select(self):
        while True:
            print()
            n = int(input('추가 주문은 음료번호를 지불은 0을 입력해주세요 : '))
            if n > 0 and n < len(self.menu_temp) + 1:
                t = 0 
                while t != 1 and t != 2:
                    t= int(input('HOT 음료는 1을, ICE 음료는 2를 입력하세요 : '))
        
                    if (t==1):
                        self.temp = 'HOT'
                    elif (t==2): 
                        self.temp = 'ICE'
                    else:    
                        print('1과 2 중 하나를 입력하세요.\n')

                self.order_menu.append(self.temp + ' ' + self.menu_temp[n-1])
                self.order_price.append(self.price_temp[n-1])
                self.price_sum += self.price_temp[n-1]

                print('추가주문 음료',self.temp, self.menu_temp[n-1], self.price_temp[n-1], '원\n', '합계 : ', self.price_sum)
            else:
                if n == 0:
                    print('주문이 완료 되었습니다.')
                    print(self.order_menu, self.order_price)
                    break
                else:  
                    print('없는 메뉴입니다. 다시 주문해 주세요.')
    def pay(self):
        n = input('결제 수단을 선택해 주세요 : ')
        if n == 'cash' or n == '1':
            print('직원을 호출하겠습니다.')
        elif n == 'card' or n == '2':
            print('IC칩 방향에 맞게 카드를 꽂아주세요.')
        else:
            print('다시 결제를 시도해 주세요.')  

    def table(self):
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        print('합계 금액 :', sum(self.order_price))

        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
