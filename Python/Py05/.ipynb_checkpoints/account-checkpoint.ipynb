```python
import random

class Account:
    account_num = 0
    
    def __init__(self, depositor, balance):
        self.bank_name = "SC은행"
        self.depositor = depositor
        self.balance = balance
        self.account_number = self._generate_account_number()
        self.deposit_count =0
        self.deposit_log = []
        self.withdraw_log = []
        Account.account_num += 1

    def _generate_account_number(self):
        part1 = random.randint(100, 999)
        part2 = random.randint(10, 99)
        part3 = random.randint(100000, 999999)
        return f"{part1}-{part2}-{part3}"   
    
    def get_account_num(self):
        print(f"생성된 계좌 수 : {Account.account_num}")

    def deposit(self, amount):
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return
            
        self.balance += amount
        self.deposit_count += 1
        self.deposit_log.append(amount)
        
        print(f"{amount}원 입금 완료. 현재 잔고: {self.balance}원")
        
        if self.deposit_count % 5 == 0:
            balance_interest = int(self.balance * 0.01)
            self.balance += balance_interest
            print(f"이자 지급 +{balance_interest}원 , 잔고: {self.balance}원")
            
    def withdraw(self, amount):
        if amount > self.balance:
            print(f"잔고 부족. 현재 잔고: {self.balance}원")
            return
            
        self.balance -= amount
        self.withdraw_log.append(amount)
        print(f"{amount}원 출금 완료. 현재 잔고: {self.balance}원")

    def display_info(self):
        print(f"(은행이름: {self.bank_name}, 예금주: {self.depositor}, 계좌번호: {self.account_number}, 잔고: {self.balance}원)")

    def deposit_history(self):
        print(f"[{self.depositor}] 입금 내역:")
        for i, amount in enumerate(self.deposit_log, 1):
            print(f"  {i}. {amount}원")

    def withdraw_history(self):
        print(f"[{self.depositor}] 출금 내역:")
        for i, amount in enumerate(self.withdraw_log, 1):
            print(f"  {i}. {amount}원")
          

accounts = [Account("파이썬", 500000), Account("아이펠", 1500000), Account("모두연", 2000000)]

accounts[0].get_account_num()

print("\n--- 잔고 100만원 이상 고객 ---")
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()

print("\n--- 입출금 테스트 ---")
for i in accounts:
    i.deposit(10000)
    i.deposit(20000)
    i.deposit(30000)
    i.deposit(40000)
    i.deposit(50000)
    i.withdraw(2000)
    i.deposit_history()
    i.withdraw_history()

```
